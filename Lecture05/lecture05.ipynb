````python
maze = """
###################..
#.......#.#.........#
#.#####.#.#######.#.#
#...#...#.#...#...#.#
#.###.###.###.###.###
#.#.#...#...#.#.....#
###.###.#.#.#.#.###.#
#...#.#...#...#...#.#
#.#.#.#.#.#.#####.###
#.#.#...#.#.#...#...#
#.#######.###.#####.#
#.............#.....#
#.###########.###.#.#
#.........#.....#.#.#
#.#.#.###.#.###.###.#
#.#.#.#.#...#.#.#...#
#.#####.#.###.#.#.#.#
#.#...#.......#...#.#
#.#.#.###.#####.#####
#...#.......#.......#
..###################
"""
````

````python
def parse(maze):
    result = []
    for line in maze.splitlines():
        line = line.strip()
        if line == '':
            continue
        result.append(list(line))
        #print(len(result[-1]))
    return result

def display(maze):
    result = ""
    for y, row in enumerate(maze):
        for x, v in enumerate(row):
            if type(v) == int:
                result += '\u001b[31m%d\u001b[0m' % (v % 10)
            elif v == "#":
                result += '#'
            else:
                result += ' '
        result += '\n'
    print(result)
    return

m = parse(maze)
display(m)
````

# Поиск в глубину

В данном подходе, мы стараемся уходить в глубь графа так сильно, на сколько это возможно

<img src="https://upload.wikimedia.org/wikipedia/commons/7/7f/Depth-First-Search.gif">

````python
from copy import deepcopy

def DepthFirst(pos, maze):
    maze = deepcopy(maze)
    x, y = pos
    assert maze[y][x] == '.'
    maze[y][x] = 0
    
    directions = [
        ( 0,  1),
        ( 0, -1),
        ( 1,  0),
        (-1,  0),
    ][::-1]
    
    queue = [pos]
    while queue:
        x, y = queue.pop()
        v = maze[y][x]
        
        for dx, dy in directions:
            if not(0 <= x + dx < len(maze[0])):
                continue
            if not(0 <= y + dy < len(maze)):
                continue
                
            xc = x + dx
            yc = y + dy
            vc = maze[yc][xc]
            if type(vc) == int:
                continue
            
            if vc == '.':
                queue.append((xc, yc))
                maze[yc][xc] = v + 1
        
        
    display(maze)
    return maze
            
solution = DepthFirst((9, 9), m)
````

````python
def ExtractPath(maze):
    N, M = len(maze[0]), len(maze)
    
    #x, y = N-1, 0
    x, y = 0, M-1
    v = maze[y][x]
    path = [(x, y)]
    
    directions = [
        ( 0,  1),
        ( 0, -1),
        ( 1,  0),
        (-1,  0),
    ]
    
    while v:
        for dx, dy in directions:
            if not(0 <= x + dx < N):
                continue
            if not(0 <= y + dy < M):
                continue
            
            vc = maze[y + dy][x + dx]
            if vc == v - 1:
                x += dx
                y += dy
                v -= 1
                path.append((x, y))
                break
        else:
            raise Exception()
    return path[::-1]

r = ExtractPath(solution)
tmp = deepcopy(m)
for step, (x, y) in enumerate(r):
    tmp[y][x] = step
display(tmp)
````

````python
from copy import deepcopy

def DepthFirstRecursive(pos, maze, prev=-1):
    x, y = pos
    v = maze[y][x]
    assert v == '.'
    maze[y][x] = prev + 1
    
    directions = [
        ( 0,  1),
        ( 0, -1),
        ( 1,  0),
        (-1,  0),
    ]
    
    for dx, dy in directions:
        if not(0 <= x + dx < len(maze[0])):
            continue
        if not(0 <= y + dy < len(maze)):
            continue

        xc = x + dx
        yc = y + dy
        vc = maze[yc][xc]
        if type(vc) == int: continue
        if vc == '#': continue
        
        DepthFirstRecursive((x + dx, y + dy), maze, prev + 1)
        
tmp = deepcopy(m)
DepthFirstRecursive((9, 9), tmp)
display(tmp)

r = ExtractPath(tmp)
tmp = deepcopy(m)
for step, (x, y) in enumerate(r):
    tmp[y][x] = step
display(tmp)
````

## Поиск в ширину

В данном варианте, мы ищем в ширь. Мы осматриваем все доступные вершины, прежде чем углублятся в граф.

<img src="https://upload.wikimedia.org/wikipedia/commons/5/5d/Breadth-First-Search-Algorithm.gif">

````python
from copy import deepcopy

def WidthFirst(pos, maze):
    maze = deepcopy(maze)
    x, y = pos
    assert maze[y][x] == '.'
    maze[y][x] = 0
    
    directions = [
        ( 0,  1),
        ( 0, -1),
        ( 1,  0),
        (-1,  0),
    ][::-1]
    
    queue = [pos]
    while queue:
        x, y = queue.pop(0)
        v = maze[y][x]
        
        for dx, dy in directions:
            if not(0 <= x + dx < len(maze[0])):
                continue
            if not(0 <= y + dy < len(maze)):
                continue
                
            xc = x + dx
            yc = y + dy
            vc = maze[yc][xc]
            if type(vc) == int:
                continue
            
            if vc == '.':
                queue.append((xc, yc))
                maze[yc][xc] = v + 1
        
        
    display(maze)
    return maze
            
solution = WidthFirst((9, 9), m)
r = ExtractPath(solution)
tmp = deepcopy(m)
for step, (x, y) in enumerate(r):
    tmp[y][x] = step
display(tmp)
````
